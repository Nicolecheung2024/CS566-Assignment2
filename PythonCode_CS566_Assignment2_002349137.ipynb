```python
#Problem 2.1

def ac3(csp):
    # Initialize the queue with all the arcs in the CSP
    queue = [(xi, xj) for xi in csp['variables'] for xj in csp['neighbors'][xi]]

    # Process each arc in the queue
    while queue:
        (xi, xj) = queue.pop(0)
        if revise(csp, xi, xj):
            if not csp['domains'][xi]:
                return False  # No values left for xi, problem is unsolvable
            for xk in csp['neighbors'][xi]:
                if xk != xj:
                    queue.append((xk, xi))
    return True

def revise(csp, xi, xj):
    revised = False
    for x in set(csp['domains'][xi]):
        if not any([x != y for y in csp['domains'][xj]]):
            csp['domains'][xi].remove(x)
            revised = True
    return revised

# Define the CSP problem
csp = {
    'variables': ['V1', 'V2', 'V3', 'V4'],
    'domains': {
        'V1': ['Red', 'Blue'],
        'V2': ['Blue', 'Green'],
        'V3': ['Blue', 'Green'],
        'V4': ['Red']
    },
    'neighbors': {
        'V1': ['V2', 'V3', 'V4'],
        'V2': ['V1', 'V3'],
        'V3': ['V1', 'V2', 'V4'],
        'V4': ['V1', 'V3']
    }
}

# Apply AC-3 algorithm to the CSP problem
result = ac3(csp)

# Print the result
print("AC-3 Algorithm Result:")
if result:
    print("The CSP is consistent. The domains after applying AC-3 are:")
    for variable in csp['variables']:
        print(f"{variable}: {csp['domains'][variable]}")
else:
    print("The CSP is unsolvable.")

```

```python
#Problem 3
from constraint import Problem

# Define the music pieces with their languages and styles
pieces = {
    1: {"language": "English", "style": "rock"},
    2: {"language": "English", "style": "rock"},
    3: {"language": "English", "style": "jazz"},
    4: {"language": "English", "style": "techno"},
    5: {"language": "English", "style": "techno"},
    6: {"language": "English", "style": "alternative"},
    7: {"language": "French", "style": "rock"},
    8: {"language": "French", "style": "jazz"},
    9: {"language": "French", "style": "techno"},
    10: {"language": "French", "style": "alternative"}
}

# Define a function to enforce the constraints
def enforce_constraints(*args):
    selected_pieces = args[:5]
    languages = [pieces[p]["language"] for p in selected_pieces]
    styles = [pieces[p]["style"] for p in selected_pieces]

    # Check consecutive language constraint
    for i in range(4):
        if languages[i] == languages[i + 1]:
            return False

    # Check consecutive style constraint
    for i in range(4):
        if styles[i] == styles[i + 1]:
            return False

    # Check for style diversity
    if not ("rock" in styles and "jazz" in styles and "techno" in styles and "alternative" in styles):
        return False

    return True

def solve_playlist_problem():
    problem = Problem()
    
    # Variables: P_i representing the selected pieces
    problem.addVariables(range(5), list(pieces.keys()))

    # Constraint: Each piece must be unique
    problem.addConstraint(lambda a, b, c, d, e: len(set([a, b, c, d, e])) == 5, range(5))

    # Constraint: Must include M10
    problem.addConstraint(lambda a, b, c, d, e: 10 in [a, b, c, d, e], range(5))

    # Constraint: Must end with a jazz piece
    problem.addConstraint(lambda a: pieces[a]["style"] == "jazz", [4])

    # Add custom constraint
    problem.addConstraint(enforce_constraints, range(5))
    
    solutions = problem.getSolutions()
    
    if solutions:
        for solution in solutions:
            playlist = [f'M{solution[i]}' for i in range(5)]
            print("Solution:", playlist)
    else:
        print("No solution found.")

# Call the function to solve the problem
solve_playlist_problem()
```
